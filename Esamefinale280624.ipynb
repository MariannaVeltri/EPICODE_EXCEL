```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv('owid-covid-data.csv')

df['date'] =pd.to_datetime(df['date']) #converti la colonna 'date' in datatime
filtered_df = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)]

x_value = filtered_df['total_cases']
y_value = filtered_df['date']

plt.plot(x_value,y_value)
plt.xlabel ('Anno di riferimento 2022')
plt.ylabel('Casi totali in Italia')
plt.title('Evoluzione dei casi totali nell\'anno 2022')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv('owid-covid-data.csv')

df['date'] = pd.to_datetime(df['date']) #converti la colonna 'date' in datatime
filtered_df = df[df['date'].dt.year == 2022]

grouped_df= filtered_df.groupby('date')['new_cases'].sum().reset_index()

x_value = grouped_df['date']
y_value = grouped_df['new_cases']

plt.plot(x_value,y_value)
plt.xlabel ('anno')
plt.ylabel('nuovi casi')
plt.title('somma nuovi casi per anno')
plt.show()
```
